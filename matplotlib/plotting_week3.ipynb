## Animation

```python
import matplotlib.animation as animation

n = 100
x = np.random.randn(n)
```

```python
def update(curr):
    if curr == n:
        a.event_source.stop()
    plt.cla()
    bins = np.arange(-4, 4, 0.5)
    plt.hist(x[:curr], bins=bins)
    plt.axis([-4, 4, 0, 30])
    plt.gca().set_title('Sampling the normal distribution')
    plt.gca().set_ylabel('Frequency')
    plt.gca().set_xlabel('Value')
    plt.annotate('n = {}'.format(curr), [3,27])
```

```python
fig = plt.figure()
a = animation.FuncAnimation(fig, update, interval=10)
```

## Interactivity

```python
plt.figure()

data = np.random.randn(10)
plt.plot(data)

def onclick(event):
    plt.cla()
    plt.plot(data)
    plt.gca().set_title('Event at pixel {}, {} {} and data {},{}'.format(
        event.x, event.y,'\n', event.xdata, event.ydata))
    
plt.gcf().canvas.mpl_connect('button_press_event', onclick)
```

```python
from random import shuffle
import pandas as pd

origins = ['China', 'Brazil', 'India', 'USA', 'France', 'England', 'Germany', 'Mexico', 'Iraq', 'Chile']
shuffle(origins)

df = pd.DataFrame({'height' : np.random.rand(10),
                  'weight' : np.random.rand(10),
                  'origin' : origins})

df
```

```python
plt.figure()
plt.scatter(df['height'], df['weight'], picker=5, label='country origin')
plt.gca().set_ylabel('Weight')
plt.gca().set_xlabel('Height')
plt.legend()
```

```python
def onpick(event):
    origin = df.iloc[event.ind[0]]['origin']
    plt.gca().set_title('Picked item came from '+event)

plt.gcf().canvas.mpl_connect('pick_event', onpick)
```
